```python
import rawpy
import imageio
import numpy as np
from IPython.display import Image

red_scale=1
green_scale=1
blue_scale=1

raw = rawpy.imread('MWW_3388.NEF')
rgb = raw.postprocess() #gamma=(1,1), no_auto_bright=True, output_bps=16)

# Adjust color tones
rgb[:, :, 0] = np.clip(rgb[:, :, 0] * red_scale, 0, 255)  # Red channel
rgb[:, :, 1] = np.clip(rgb[:, :, 1] * green_scale, 0, 255)  # Green channel
rgb[:, :, 2] = np.clip(rgb[:, :, 2] * blue_scale, 0, 255)  # Blue channel

# Save the adjusted image
imageio.imsave('MWW_3388.JPG', rgb.astype(np.uint8))

#imageio.imsave('MWW_3388.JPG', rgb)
Image(filename='MWW_3388.JPG') 
```
